```python
# Import Necessary Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Train data
data = pd.read_csv("data/train_FD001.txt", sep=' ', header = None)
data
```

```python
# Drop NaN values columns
data = data.drop([26, 27], axis = 'columns')
data
```

```python
# 3.2 Add column names
data.columns = ['unit_number', 'cycle', 'setting1', 'setting2', 'setting3', 'sensor1',
                    'sensor2', 'sensor3', 'sensor4', 'sensor5', 'sensor6', 'sensor7', 'sensor8',
                    'sensor9', 'sensor10', 'sensor11', 'sensor12', 'sensor13', 'sensor14',
                    'sensor15', 'sensor16', 'sensor17', 'sensor18', 'sensor19', 'sensor20', 'sensor21']
data.head()
```

RUL = EOL - Cycle
* RUL = Remaining Useful Life
* EOL = End of Life in Cycle

```python
# 3.5 Create an array which consists of End of Life (EOL) values for each units
EOL = []
for each_unit in data['unit_number']:
    eol_values = ((data[data['unit_number'] == each_unit]["cycle"]).values[-1])
    EOL.append(eol_values)
```

```python
# 3.6 Append "EOL" in a new column
data["EOL"] = EOL
data
```

```python
data["RUL"] = data["EOL"] - data["cycle"]
new_data = data.drop(columns=["EOL"])
new_data
```

## Feature analysis

```python
# Let's check the statistical description of each feature columns
new_data.describe()
```

```python
# Let's check how many uniques values are presented in each columns
new_data.nunique(axis = 0)
```

#### Box Plot

```python
plt.figure(figsize = (15, 21))
for i in np.arange(2, 26):
    temp = new_data.iloc[:, i]
    plt.subplot(5, 6, i)
    plt.boxplot(temp)
    plt.title(new_data.columns[i])
    
plt.show()
```

#### Correlation Heat-Map

```python
corrmap = new_data.corr()
top_corr_features = corrmap.index
plt.figure(figsize = (58, 58))
# plt.savefig('Corrmap.png, bbox_inches = 'tight)
sns.set(font_scale = 4, font = "Times New Roman")

# Plot heatmap
g=sns.heatmap(new_data[top_corr_features].corr(),cmap="RdYlGn", linewidths=0.1, annot=True, annot_kws={"size":35})
g.set_xticklabels(g.get_xmajorticklabels(), fontsize = 35)
g.set_yticklabels(g.get_xmajorticklabels(), fontsize = 35)
```

#### Drop columns which have less than 0.5 correlation with RUL

```python
delete_columns=[]

for col in new_data.columns:
    corr = new_data[col].corr(new_data['RUL'])
    #print(col,corr)
    if abs(corr)>0.5:
        pass
    else:    
        delete_columns.append(col)
        
delete_columns
```

```python
final_data = new_data.drop(columns=delete_columns)
final_data
```

#### Check Correlation Heat-Map of final data

```python
corrmat = final_data.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(58,58))
sns.set(font_scale=4,font="Times New Roman")

#plot heat map
g=sns.heatmap(final_data[top_corr_features].corr(),cmap="RdYlGn", linewidths=0.1, annot=True, annot_kws={"size":35})
g.set_xticklabels(g.get_xmajorticklabels(), fontsize = 35)
g.set_yticklabels(g.get_xmajorticklabels(), fontsize = 35)
```

### Extra Tree Regession for Feature Importance

```python
X = final_data.iloc[:, 0:13]
y = final_data.iloc[:, 13]


from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(X,y)
print(model.feature_importances_)
```

## Plot

```python
feat_importances = pd.Series(model.feature_importances_,index = X.columns)
feat_importances=feat_importances.sort_values( ascending=False)
feat_importances.nlargest(13).plot(kind='barh',fontsize =13)
```
