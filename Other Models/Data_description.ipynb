### Turbo Engine Data Description

#### 1 Import Libraries

```python
# Import necessary Libraries
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score
```

#### 2 Load Raw Data

```python
# Train data
raw_data = pd.read_csv("data/train_FD001.txt", sep=' ', header = None)
raw_data.head()
```

#### 3 Preprocess raw data

```python
#3.1 Drop features or columns with NaN value
raw_data = raw_data.drop([26, 27], axis = 'columns')
raw_data.head()
```

```python
# 3.2 Add column names
raw_data.columns = ['unit_number', 'cycle', 'setting1', 'setting2', 'setting3', 'sensor1',
                    'sensor2', 'sensor3', 'sensor4', 'sensor5', 'sensor6', 'sensor7', 'sensor8',
                    'sensor9', 'sensor10', 'sensor11', 'sensor12', 'sensor13', 'sensor14',
                    'sensor15', 'sensor16', 'sensor17', 'sensor18', 'sensor19', 'sensor20', 'sensor21']
raw_data.head()
```

```python
# 3.3 Let's check any specific unit values, for example let consider unit 1
raw_data[raw_data['unit_number'] == 1]
```

```python
# 3.4 How to extract last cycle value of each engine unit
```

```python
x1 = raw_data[raw_data['unit_number'] == 1]
x2 = (x1["cycle"]).values
print(x2)
print(x2[-1])
```

```python
last_cycle_value = ((raw_data[raw_data['unit_number'] == 1]["cycle"]).values[-1])
print(last_cycle_value)
```

```python
# 3.5 Create an array which consists of End of Life (EOL) values for each units
EOL = []
for each_unit in raw_data['unit_number']:
    eol_values = ((raw_data[raw_data['unit_number'] == each_unit]["cycle"]).values[-1])
    EOL.append(eol_values)
    
print(EOL)
```

```python
# 3.6 Append "EOL" in a new column
raw_data["EOL"] = EOL
raw_data
```
